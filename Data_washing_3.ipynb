```python
def plot_force_sensors(data, title='', save_path=None):
    """
    Y/Z2x2
    
    :
        data: numpy[n,16][FS6_Y, FS6_Z, FS2_Y, FS2_Z, FS4_Y, FS4_Z, FS3_Y, FS3_Z, ...]
        title:  (: '')
        save_path:  ()
    """
    # 
    plt.rcParams['font.sans-serif'] = ['SimSun']
    plt.rcParams['axes.unicode_minus'] = False
    
    # 
    if data.shape[1] != 8:
        raise ValueError("86 (42)")
    
    # 2x2
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))
    fig.suptitle(title, fontsize=16)

    #  (, , )
    sensor_config = [
        ('FS2', axs[0, 1], (0, 1)),   # : FS2 (Y:0, Z:1)
        ('FS6', axs[0, 0], (2, 3)),   # : FS6 (Y:2, Z:3)
        ('FS4', axs[1, 0], (4, 5)),   # : FS4 (Y:4, Z:5)
        ('FS3', axs[1, 1], (6, 7))    # : FS3 (Y:6, Z:7)
    ]

    # 
    for name, ax, (y_col, z_col) in sensor_config:
        y_data = data[:, y_col]
        z_data = data[:, z_col]
        time = np.arange(len(y_data))
        
        ax.plot(time, y_data, 'b-', label='Z')
        ax.plot(time, z_data, 'r-', label='Y')
        ax.set_title(name)
        ax.set_xlabel('')
        ax.set_ylabel('(N)')
        ax.set_ylim(-4, 4)
        ax.legend()
        ax.grid(True, linestyle='--', alpha=0.6)

    # 
    plt.tight_layout()
    plt.subplots_adjust(top=0.92)

    # 
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f": {save_path}")
    
    plt.show()
```

```python
def force_washing(data):
    data_swaped = swap_columns(data)
    for i in range(4):
        # force in leg 1 and 3
        if i % 2 == 0:
            data_swaped[:,32+4*i:36+4*i] -= calculate_column_averages(data_swaped[:,32+4*i:36+4*i],125,135)
        else:# force in leg 2 and 4
            data_swaped[:,32+4*i:36+4*i]-=calculate_column_averages(data_swaped[:,32+4*i:36+4*i],75,85 )
    return data_swaped
```

```python
def calculate_column_averages(data, x, y):
    """
    (n,16)xy(16,)
    
    :
    data -- (n, 16)numpy
    x -- ()
    y -- ()
    
    :
    (16,)numpyxy
    """
    # xy
    if x < 0 or y >= data.shape[0] or x > y:
        raise ValueError("xy")
    
    # xy
    selected_rows = data[x:y+1, :]
    
    # 
    column_averages = np.mean(selected_rows, axis=0)
    
    return column_averages
```

```python
''''''
def get_data_in_directory(path):
    data = []
    for file_name in os.listdir(path):
        file_path = os.path.join(path, file_name)
        
        # 
        data.append(np.load(file_path).T)
    return data
```

```python
def swap_columns(data):
    """
    [36:40][40:44]
    
    :
    data -- (n, 54)numpy
    
    :
    
    """
    # 
    swapped_data = data.copy()
    
    # [36:40][40:44]
    swapped_data[:, 36:40], swapped_data[:, 40:44] = data[:, 40:44], data[:, 36:40]
    
    return swapped_data
```

```python
def plot_leg_joints(data, joint_label='', title='', save_path=None):
    """
    2x2
    
    :
        data: numpy[n,16]
        joint_label:  (: '')
        title:  (: '')
        save_path:  ()'output.png'
    """
    # 
    plt.rcParams['font.sans-serif'] = ['SimSun']  #  'STSong'
    plt.rcParams['axes.unicode_minus'] = False  # 
    
    # 
    if data.shape[1] != 16:
        raise ValueError("16 (44)")
    
    # 2x2
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    fig.suptitle(title, fontsize=16)

    # 
    leg_names = ['1', '2', '3', '4']
    joint_names = [f'{joint_label}{i+1}' for i in range(4)]  # 
    colors = ['r', 'g', 'b', 'm']  # 
    line_styles = ['-', '--', ':', '-.']  # 

    # 
    for leg_idx in range(4):
        # 
        if leg_idx == 0:  # Leg1 -> (0,1)
            ax = axs[0, 1]
        elif leg_idx == 1:  # Leg2 -> (0,0)
            ax = axs[0, 0]
        elif leg_idx == 2:  # Leg3 -> (1,0)
            ax = axs[1, 0]
        else:  # Leg4 -> (1,1)
            ax = axs[1, 1]
        
        #  (4)
        leg_data = data[:, leg_idx*4 : (leg_idx+1)*4]
        
        # 
        for joint_idx in range(4):
            ax.plot(leg_data[:, joint_idx], 
                    color=colors[joint_idx],
                    linestyle=line_styles[joint_idx],
                    linewidth=1.5,
                    label=joint_names[joint_idx])
        
        # 
        ax.set_title(f'{leg_names[leg_idx]}{joint_label}')
        ax.set_xlabel('')
        ax.set_ylabel(f'{joint_label}(rad)')
        ax.legend(loc='upper right')
        # ax.grid(True, linestyle='--', alpha=0.6)

    # 
    plt.tight_layout()
    plt.subplots_adjust(top=0.92)

    # 
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f": {save_path}")
    plt.show()
```
